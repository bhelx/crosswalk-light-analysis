```python
import pandas as pd
import numpy as np
import json
from matplotlib import pyplot as plt
```

```python
data = json.loads(open('data/lights.json').read())
```

```python
dataframe = pd.DataFrame(data['lights'])
dataframe.set_index('_id')
dataframe.head()
```

```python
figsize(11, 9)
dataframe['status'].value_counts().plot(kind='bar', color=['#00FF00', '#FF0000', '#FFFF00'])
```

```python
figure(1, figsize=(6,6))
labels = ['Working', 'Unusable', 'Broken']
title('Status of Walking Lights in New Orleans (CBD)', bbox={'facecolor':'0.8', 'pad':5})
plt.pie(dataframe['status'].value_counts(normalize=True).values, labels=labels, colors=['#00FF00', '#FF0000', '#FFFF00'], autopct='%1.1f%%')
```
